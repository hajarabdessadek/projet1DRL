```python
#exercice2
import gymnasium as gym
env = gym.make('CartPole-v1', render_mode='human')
num_episodes = 5  
all_observations = []  
all_rewards = []        

for episode in range(num_episodes):  
    state, _ = env.reset()
    episode_rewards = 0 
    episode_observations = []  
    step_count = 0
    done = False  

    print(f"\n Épisode {episode + 1} ")

    while not done:  
        action = env.action_space.sample()  
        next_state, reward, terminated, truncated, _ = env.step(action)
        done = terminated or truncated  
        episode_observations.append(next_state)
        episode_rewards += reward

        print(f"  Étape {step_count + 1}:")
        print(f"    Action choisie: {action}")
        print(f"    Observation: {next_state}")
        print(f"    Récompense: {reward}")
        print(f"    Terminé: {done}")

        step_count += 1
        env.render()

    all_observations.append(episode_observations)
    all_rewards.append(episode_rewards)

env.close()

print("\nAnalyse des essais")
for i, rewards in enumerate(all_rewards):
    print(f"Épisode {i + 1}: Récompense totale = {rewards}")

print("\n Variations des observations ")
for i, observations in enumerate(all_observations):
    print(f"\nÉpisode {i + 1}:")
    for step, observation in enumerate(observations):
        print(f"  Étape {step + 1}: Observation = {observation}")
```

```python
#exerciece3
import gymnasium as gym


env = gym.make('CartPole-v1', render_mode='human')


state, _ = env.reset()
episode_duration = 0  


print("Contrôle manuel de l'agent :")
print("Entrez 0 pour action 0 ou 1 pour action 1.")

while True:
    
    print(f"État actuel : {state}")

    action = int(input("Entrez une action (0 ou 1) : "))

    
    if action not in [0, 1]:
        print("Action invalide. Entrez 0 ou 1.")
        continue

   
    next_state, reward, done, truncated, info = env.step(action)

    print(f"Action : {action}")
    print(f"Récompense obtenue : {reward}")
    print(f"État suivant : {next_state}")
    
    state = next_state
    episode_duration += 1  

    
    env.render()


    if done or truncated:
        print(f"\nL'épisode est terminé après {episode_duration} étapes.")
        break


env.close()

```

```python
#exercice4
import gymnasium as gym
import numpy as np


env = gym.make('CartPole-v1')
num_episodes = 10 
episode_durations = []

for episode in range(num_episodes):
    state, _ = env.reset()
    episode_duration = 0  
    while True:
        action = env.action_space.sample()
        
        next_state, reward, done, truncated, info = env.step(action)
        
        state = next_state
        episode_duration += 1 

        if done or truncated:
            break

    episode_durations.append(episode_duration)

average_duration = np.mean(episode_durations)


print(f"Durée moyenne des épisodes (sur {num_episodes} épisodes) : {average_duration:.2f} étapes")

print("Durées des épisodes :")
for i, duration in enumerate(episode_durations):
    print(f"Épisode {i + 1}: {duration} étapes")


env.close()
```
